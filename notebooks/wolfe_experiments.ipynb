### `Test 2`

```python
configs = [['report', 'opinion', '../gispy/wolfe_gispy.csv'],
           ['method', 'discussion', '../gispy/wolfe_gispy.csv'],
           ['no', 'yes', '../gispy/disney_gispy.csv']]

all_best_combos = list()

for config in configs:
    print("*** config: {}".format(str(config)))
    exp_result = {'dev': [[],[],[]],
                  'test': [[],[],[]]}

    for seed in [42, 107, 137]:
        kfolds = create_kfolds(config[:2], config[2], n_folds=2, split_seed=seed)
        for (train, test) in kfolds:
            prefix_names = dict()
            prefix_names['gist_no'] = config[0]
            prefix_names['gist_yes'] = config[1]
            results = Wolfe().run_benchmark(config[2], 
                                            prefix_names, 
                                            train, 
                                            use_wolfe_vars=False, 
                                            use_gispy_vars=True, 
                                            custom_vars=[], 
                                            plot=False)
            results = results.head(1)
            
            all_best_combos.append(results)
            
            for idx, row in results.iterrows():
                combo = str(row['vars_name'])
                exp_result['dev'][0].append(row['mean_gist_yes'])
                exp_result['dev'][1].append(row['mean_gist_no'])
                exp_result['dev'][2].append(row['ttest_pvalue'])

                combo = combo.split('#')

                test_results = Wolfe().run_benchmark(config[2], 
                                                     prefix_names, 
                                                     test, 
                                                     use_wolfe_vars=False, 
                                                     use_gispy_vars=True, 
                                                     custom_vars=[combo[1], combo[3], combo[4], combo[5], combo[6]], 
                                                     plot=False)
                exp_result['test'][0].append(test_results.iloc[0]['mean_gist_yes'])
                exp_result['test'][1].append(test_results.iloc[0]['mean_gist_no'])
                exp_result['test'][2].append(test_results.iloc[0]['ttest_pvalue'])

        print('done processing seed {}'.format(seed))

    for k, v in exp_result.items():
        print('\n*** {} ***'.format(k))
        print('yes -> mean: {}, stdev: {}'.format(statistics.mean(v[0]), statistics.stdev(v[0])))
        print('no -> mean: {}, stdev: {}'.format(statistics.mean(v[1]), statistics.stdev(v[1])))
        print('ttest_pvalue -> mean: {}, stdev: {}'.format(statistics.mean(v[2]), statistics.stdev(v[2])))
    
    print('\n')
```

```python
for combo in all_best_combos:
    for idx, row in combo.iterrows():
        indices = row['vars_name'].split('#')
        print("{}, {}, {}, {}, {}".format(indices[1], indices[3], indices[4], indices[5], indices[6]))
```

There are two types of text for each disease: 1) `Patient` version, and 2) `Health Professional (HP)` version. And for each type, there are different texts separately annotated for gist and verbatim (T1, T2, T3, etc.) Based on Chris Wolfe: "the articles for Physicians (HP, I guess) tended to be more gist whereas the ones for patients tended to be more verbatim"

For now, since there are no detailed scores available for different texts, we don’t use the NCI-related texts for evaluating our tool.
